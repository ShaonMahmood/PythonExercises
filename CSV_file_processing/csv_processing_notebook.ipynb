```python
import numpy as np
import pandas as pd
```

```python
# download the csv file content from url and convert it to pandas dataframe
csv_data_url = 'https://press.radiocut.fm/bicycle-travels-jan-2018.csv'
df = pd.read_csv(csv_data_url, encoding='utf-8')
```

```python
df.pickup_time = pd.to_datetime(df.pickup_time)
df.set_index(pd.DatetimeIndex(df.pickup_time),inplace=True)
morning_trips = df.between_time('6:00AM','12:00PM')
```

```python
'''
What is this trickery with .to_frame() and .reset_index()? 
Because we want to sort by our calculated field (size), 
this field needs to become part of the DataFrame
'''
top5 = morning_trips.groupby(['source_station_name','source_station_code']).size().to_frame('size').reset_index().sort_values(['size'],ascending=False).head()
```

```python
top5
```

```python
def print_top_5_docking_stations(topdf):
    print('Top five source stations:\n')
    for index,row in topdf.iterrows():
        print('{0} (code: {1}): {2} trips'.format(row['source_station_name'], row['source_station_code'], row['size']))
```

```python
print_top_5_docking_stations(top5)
```
